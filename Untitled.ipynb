```python
import json
import pandas
import matplotlib
import tabulate
from pprint import pprint

restaurants = json.loads(open("restaurants.json").read())
reward = 'maxReward'

```

```python
def by_discount(elem):
  return elem[reward]
restaurants.sort(key=by_discount, reverse=True)
```

```python
df = pandas.DataFrame(restaurants)
grouped = df.groupby(reward)
discount_group = grouped.groups
```

```python
yelp_groups = df.groupby('yelpRating')

yelp_stats = yelp_groups[reward].agg(['max', 'min', 'count', 'median', 'mean'])
print(yelp_stats)
```

```python
yelp_stats.reset_index().plot.scatter(x='yelpRating',
                    y='mean',
                 c='DarkBlue')
```

```python
yelp_groups[reward].agg(['max', 'min', 'count', 'median', 'mean']).
```

```python
df[reward].hist(figsize=(15, 3),bins=20)
```

```python
print("Discount distribution")
for k, v in discount_group.items():
  print(f"{k}: {len(v)}")
  print(v)
  print(f"{k}: {discount_group[v].mean()}")
```

```python
bins = pandas.cut(df[reward], [0, 5, 10,15,20,25,30,35,40,45,50])
agg = df.groupby(bins)[reward].agg(['count'])
print(agg.to_markdown())
```

```python
agg.plot(drawstyle="steps")
```

```python
print("Top discount restaurants")
for i in range(10):
  name = restaurants[i]['name']
  print(f"{name}: {restaurants[i][reward]}%")

for r in restaurants:
  if 'Boulton' in r['name']:
    print(r)
```

```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
import os
import geoplot
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
```

```python

df = pandas.DataFrame(restaurants)

geometry = [Point(xy) for xy in zip(df['longitude'], df['latitude'])]
gdf = GeoDataFrame(df, geometry=geometry)   
```

```python
print(os.path.abspath('nyc.zip'))
```

```python
#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('nybb'))
gdf.plot(ax=world.plot(figsize=(10, 6)))
```

```python
crs={'init':'epsg:4326'}
boroughs = gpd.read_file(geoplot.datasets.get_path('nyc_boroughs'))
geo_df=gpd.GeoDataFrame(df,crs=crs,geometry=gpd.points_from_xy(df["longitude"], df["latitude"]))
```

```python
les_df = geo_df[geo_df['latitude'].between(40.7,40.767645)]
les_df[['name',reward]].head(10)
```

```python
fig,ax=plt.subplots(figsize=(15,15))
boroughs.plot(ax=ax,alpha=0.4,color="grey")
les_df.plot(ax=ax,markersize=2, alpha=0.4,color="green", label="Restaurants on Seated")
plt.show
```

```python
fig,ax=plt.subplots(figsize=(15,15))
boroughs.plot(ax=ax,alpha=0.4,color="grey")
geo_df[geo_df['latitude'] < 41].plot(ax=ax,markersize=2, alpha=0.4,color="green", label="Restaurants on Seated")
plt.savefig('seated.png',dpi=300)
```
